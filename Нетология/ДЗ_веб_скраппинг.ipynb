```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def habr_search(queries):
    base_url = "https://habr.com"
    search_url = f"{base_url}/ru/search/"
    
    results = []


    for query in queries:
        response = requests.get(search_url, params={'q': query})
      

        soup = BeautifulSoup(response.text, 'html.parser')
        articles = soup.find_all('article', class_='tm-articles-list__item')
        
        for article in articles:
            title_element = article.find('a', class_='tm-title__link')
            date_element = article.find('time')
            link_element = title_element['href'] if title_element else None

            if title_element and date_element and link_element:
                title = title_element.text.strip()
                date = date_element['datetime'].strip()
                link = base_url + link_element.strip()
                
                results.append((date, title, link))

    df = pd.DataFrame(results, columns=['Дата', 'Заголовок', 'Ссылка'])
    
    return df

# Пример использования функции
queries = ['Pandas']
df = habr_search(queries)
print(df)
```
