```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
# пункт 1

def result_p():
    
    ''' возвращает имя по номеру документа'''
    number = input('Введите номер документа: ')
    for_search = {dicts['number']: dicts['name'] for dicts in documents}
    if number in for_search.keys():
        return f'Владелец документа: {for_search[number]}'
    else:
        return f'Документ не найден в базе'
```

```python
# пункт 2

def result_s():
    
    ''' возвращает полку по номеру документа'''
    number = input('Введите номер документа: ')
    found = [i for i, j in directories.items() if number in j]
    
    if found == []:
        return f'Документ не найден в базе'
    else:
        return f'{found[0]}'
```

```python
# пункт 3

def result_l():
    
    ''' Вся информация по всем документам'''
    for doc in documents:
        for sh, docs in directories.items():
            if doc["number"] in docs:
                shelf = sh
        print(f'№ {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {shelf}')
```

```python
# пункт 4

def result_ads():
    
    '''добавляет новую полку'''
    n = input('Введите номер полки ')
    
    if n in directories.keys():
        print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(list(directories.keys()))}')
    else:
        directories[n] = []
        print(f'Полка {n} добавлена. Текущий перечень полок: {", ".join(list(directories.keys()))}')
```

```python
# пункт 5

def result_ds():
    '''удаляет полку, если она пустая'''
    
    n = input('Введите номер полки ')
    
    if n in directories.keys() and directories[n] == []:
        del directories[n]
        print(f'Полка {n} удалена. {", ".join(list(directories.keys()))}')
    
    elif n in directories.keys() and directories[n] != []:   
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(list(directories.keys()))}')
    
    else:
        print(f'Такой полки ({n}) не существует {", ".join(list(directories.keys()))}')
```

```python
# пункт 1 *

def result_ad():
    
    '''добваляет новый документ в данные'''
    
    number = input('Введите номер документа: ')
    type_doc = input('Введите тип документа: ')
    owner = input('Введите владельца документа: ')
    shelf = input('Введите полку для хранения: ')
    
    if shelf in directories.keys():
        documents.append({'type': type_doc, 'number': number, 'name': owner})
        directories[shelf] += [number]
        print(f'Документ добавлен. Текущий список документов:')
        result_l()
    else:
        print(f'Такой полки ({shelf}) не существует. Добавьте полку командой ads.')
    
```

```python
# пункт 2 *

def result_d():
    ''' удаляет документ из данных '''
    
    number = input('Введите номер документа: ')
    
    if number not in [doc['number'] for doc in documents]:
        print(f'Документ {number} не найден в базе.')
        print(f'Текущий список документов:')
        result_l()
    else:
        for doc in documents:
            if doc['number'] == number:
                documents.remove(doc)
        
        for shelf in directories:
            if number in directories[shelf]:
                directories[shelf].remove(number)
    
        print(f'Документ {number} удален.')
        print(f'Текущий список документов:')
        result_l()
```

```python
# пункт 3 *

def result_m():
    ''' перемещает документ с полки на полку '''
    
    number = input('Введите номер документа: ')
    shelf = input('Введите номер полки: ')
    
    if number not in [doc['number'] for doc in documents]:
        print(f'Документ {number} не найден в базе.')
        print(f'Текущий список документов:')
        result_l()
    
    elif shelf not in directories.keys():
        print(f'Такой полки {shelf} не существует.')
        print(f'Текущий перечень полок: {", ".join(list(directories.keys()))}')
        
    else:
        for sh, docs in directories.items():
            if number in docs:
                shelf_old = sh
        if shelf_old == shelf:
            print('Не перекладываем документ с полки на саму же себя')
        else:
            directories[shelf_old].remove(number)
            directories[shelf] += [number]
            print(f'Документ {number} перемещен')
            print(f'Текущий список документов:')
            result_l()
```

```python
def main():

    while True:
        # Ввод команд 
        
        command = input('Введите команду: ')

        if command == 'q':
            break
       
        # имя по номеру документа
        elif command == 'p':
            print(result_p())
        
        # полку по номеру документа
        elif command == 's':
            print(f'Документ хранится на полке: {result_s()}')
        
        # вся информация по всем
        elif command == 'l':
            result_l()
        
        # добавить полку
        elif command == 'ads':
            result_ads()
        
        # удалить полку, если пустая
        elif command == 'ds':
            result_ds()
        
        # добавить документ в структуру  
        elif command == 'ad':
            result_ad()
        
        # удалить документ из структуры
        elif command == 'd':
            result_d()
        
        # переместить документ с полки на полку
        elif command == 'm':
            result_m()
        
        else:
            print(f'Введена неверная команда {command}')
```

```python
main()
```
